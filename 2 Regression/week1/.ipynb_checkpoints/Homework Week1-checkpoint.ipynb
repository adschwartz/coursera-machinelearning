```python
def simple_linear_regression(input_feature, output):
    n = input_feature.size
    mean_y = output.mean()
    mean_x = input_feature.mean()
    mean_xy = (output*input_feature).mean()
    mean_x_sq = (input_feature*input_feature).mean()
    slope = (mean_xy - mean_x*mean_y/n)/(mean_x_sq-mean_x*mean_x/n)
    intercept = mean_y/n - slope*mean_x/n
    return(intercept, slope)
```

```python
intercept, slope = simple_linear_regression(sales_train['sqft_living'], sales_train['price'])
```

```python
def get_residual_sum_of_squares(input_feature, output, intercept, slope):
    rss = (output - (intercept + slope * input_feature))#.mean()
    return(rss)
```

```python
rss = get_residual_sum_of_squares(sales_train['sqft_living'], sales_train['price'], intercept, slope)
```
